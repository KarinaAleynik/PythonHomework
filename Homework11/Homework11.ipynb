```python
from sympy import*
```

```python
x = Symbol('x, y')
f1 = x ** 3 - 50 * x
f2 = -x ** 4 + 88 * x ** 2 - 241
```

```python
f1
```

```python
f2
```

```python
# Найти точки пересечения этих функций"

z = f1 - f2
```

```python
z
```

```python
x1, x2, x3, x4 = solve(z)
```

```python
x1.evalf(3), x2.evalf(3), x3.evalf(3), x4.evalf(3)
```

```python
# или
x1=x1.evalf(chop=True)
x2=x2.evalf(chop=True)
x3=x3.evalf(chop=True)
x4=x4.evalf(chop=True)
```

```python
x1, x2, x3, x4
```

```python
y1=f1.subs(x,x1)
y2=f1.subs(x,x2)
y3=f1.subs(x,x3)
y4=f1.subs(x,x4)
```

```python
y1, y2, y3, y4
```

```python
# Построить графики функций в одной системе координат
z = plot((f1), (f2), show=False)
z[0].line_color = 'red'
z[1].line_color = 'black'
z.show()
```

```python
# Построить график общей функции
h = plot((f1+f2), show=False)
h[0].line_color = 'yellow'
h.show()
```
